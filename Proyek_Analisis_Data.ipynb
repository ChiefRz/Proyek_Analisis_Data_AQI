## Exploratory Data Analysis (EDA)

### Explore Tabel data_pencemaran

```python
# Konversi tahun ke variabel kategorikal
data_pencemaran.loc[:, 'month'] = data_pencemaran['month'].astype('category')
data_pencemaran.loc[:, 'year'] = data_pencemaran['year'].astype('category')
data_pencemaran.info()

# Memeriksa nilai unik di kolom 'month' setelah penghapusan
print(data_pencemaran['month'].unique())  # Untuk memeriksa nilai unik

# Grouping dan menghitung rata-rata TEMP
data_grouped = data_pencemaran.groupby(['station', 'month', 'year'])['PM10'].mean().reset_index()
data_grouped
```

```python
# Membuat Fungsi untuk menentukan musim berdasarkan bulan
def menentukan_musim(month):
    if month in [12, 1, 2]:
        return 'Musim Dingin'
    elif month in [3, 4, 5]:
        return 'Musim Semi'
    elif month in [6, 7, 8]:
        return 'Musim Panas'
    elif month in [9, 10, 11]:
        return 'Musim Gugur'
    else:
        return 'Invalid month'

# Merapkan fungsi untuk menambahkan kolom musim
data_grouped['musim'] = data_grouped['month'].apply(menentukan_musim)

# Mampilkan DataFrame
print(data_grouped)
data_grouped
```

```python
# Membuat fungsi untuk mengkategorikan AQI
def calculate_aqi_pm10(concentration):
    # Menentukan batas bawah dan batas atas konsentrasi dan rentang AQI
    BP_low = 55
    BP_high = 154
    I_low = 51
    I_high = 100

    # Hitung AQI dengan menggunakan rumus
    AQI_PM10 = ((concentration - BP_low) / (BP_high - BP_low)) * (I_high - I_low) + I_low
    return AQI_PM10

# Menghitung AQI untuk setiap nilai di kolom 'pm10' dan menambahkannya ke kolom baru 'aqi_pm10'
data_grouped['aqi_pm10'] = data_grouped['PM10'].apply(calculate_aqi_pm10)

# Menampilkan DataFrame
print(data_grouped)
```

```python

data_grouped_aqi = data_grouped

# Konversi tahun ke variabel kategorikal
data_grouped_aqi.loc[:, 'year'] = data_grouped_aqi['year'].astype('category')

# Pastikan untuk memeriksa nilai unik di kolom 'year' setelah penghapusan
print(data_grouped_aqi['year'].unique())

# Grouping dan menghitung rata-rata TEMP
data_grouped_aqi = data_grouped_aqi.groupby(['station', 'year'])['aqi_pm10'].mean().reset_index()
data_grouped_aqi

```

```python
# Mengelompokkan AQI
def air_quality(aqi):
    if aqi <= 50:
        return 'Baik'
    elif aqi <= 100:
        return 'Sedang'
    elif aqi <= 150:
        return 'Tidak Sehat'
    elif aqi <= 200:
        return 'Sangat Tidak Sehat'
    else:
        return 'Berbahaya'

data_grouped_aqi['kualitas_udara'] = data_grouped_aqi['aqi_pm10'].apply(air_quality)

# Menampilkan DataFrame
print(data_grouped_aqi)
```

```python
# Menginisiasi data latitude dan longitude
data = {
    'station': ['Aotizhongxin', 'Changping', 'Dingling', 'Dongsi', 'Guanyuan', 'Gucheng', 'Huairou', 'Nongzhanguan', 'Shunyi', 'Tiantan', 'Wanliu', 'Wanshouxigong'],
    'latitude': [39.9092, 40.2000, 40.2350, 39.9330, 39.9510, 39.9130, 40.4300, 39.9320, 40.1300, 39.8840, 39.9790, 39.9050],
    'longitude': [116.3972, 116.2000, 116.1960, 116.4160, 116.3090, 116.2040, 116.6400, 116.4500, 116.6520, 116.4140, 116.3070, 116.3070]
}

df = pd.DataFrame(data)
print(df)
df
```

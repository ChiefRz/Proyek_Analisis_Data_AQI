## Exploratory Data Analysis (EDA)

### Explore Tabel data_pencemaran

Konversi tahun ke variabel kategorikal

```python
data_pencemaran.loc[:, 'month'] = data_pencemaran['month'].astype('category')
data_pencemaran.loc[:, 'year'] = data_pencemaran['year'].astype('category')
data_pencemaran.info()
print(data_pencemaran['month'].unique())
```

Grouping dan menghitung rata-rata TEMP

```python
data_grouped = data_pencemaran.groupby(['station', 'month', 'year'])['PM10'].mean().reset_index()
data_grouped
```

Membuat Fungsi untuk menentukan musim berdasarkan bulan

```python
def menentukan_musim(month):
    if month in [12, 1, 2]:
        return 'Musim Dingin'
    elif month in [3, 4, 5]:
        return 'Musim Semi'
    elif month in [6, 7, 8]:
        return 'Musim Panas'
    elif month in [9, 10, 11]:
        return 'Musim Gugur'
    else:
        return 'Invalid month'
```

Merapkan fungsi untuk menambahkan kolom musim

```python
data_grouped['musim'] = data_grouped['month'].apply(menentukan_musim)

data_grouped
```

Membuat fungsi untuk mengkategorikan AQI

```python
def calculate_aqi_pm10(concentration):
    BP_low = 55
    BP_high = 154
    I_low = 51
    I_high = 100

    AQI_PM10 = ((concentration - BP_low) / (BP_high - BP_low)) * (I_high - I_low) + I_low
    return AQI_PM10
```

Menghitung AQI untuk setiap nilai di kolom 'pm10' dan menambahkannya ke kolom baru 'aqi_pm10'

```python
data_grouped['aqi_pm10'] = data_grouped['PM10'].apply(calculate_aqi_pm10)

print(data_grouped)
```

Grouping dan menghitung rata-rata TEMP

```python
data_grouped_aqi = data_grouped
data_grouped_aqi = data_grouped_aqi.groupby(['station', 'year'])['aqi_pm10'].mean().reset_index()
data_grouped_aqi
```

Mengelompokkan AQI

```python
def air_quality(aqi):
    if aqi <= 50:
        return 'Baik'
    elif aqi <= 100:
        return 'Sedang'
    elif aqi <= 150:
        return 'Tidak Sehat'
    elif aqi <= 200:
        return 'Sangat Tidak Sehat'
    else:
        return 'Berbahaya'

data_grouped_aqi['kualitas_udara'] = data_grouped_aqi['aqi_pm10'].apply(air_quality)
print(data_grouped_aqi)
```

### Geospatial Analysis

Membuat DataFrame dengan menghilangkan data NaN pada kolom kualitas_udara.

```python
AQI_df = AQI_all_df.dropna(subset=['kualitas_udara'])
AQI_df
```

Menentukan Stasiun dan Tahun yang ingin di highlight

```python
nama_stasiun3 = 'Gucheng'
year3 = 2013
```

Menyaring data berdasarkan tahun dan stasiun serta mengambil lokasi geografis dari stasiun tersebut

```python
data_filtered_p = AQI_df[AQI_df['year'] == year3]
station_data = AQI_df[AQI_df['station'] == nama_stasiun3].iloc[0]
location = [station_data['latitude'], station_data['longitude']]
```

Membuat peta

```python
m = folium.Map(location=location, zoom_start=12,  tiles='CartoDB positron')
```

Menampilkan kualitas udara di berbagai stasiun pada peta dan menyediakan informasi tambahan melalui popup.

```python
for index, row in data_filtered_p.iterrows():
    if row['kualitas_udara'] == 'Baik':
        color = 'green'
    elif row['kualitas_udara'] == 'Sedang':
        color = 'orange'
    else:  # 'Buruk'
        color = 'red'

    popup_text = f"Station: {row['station']}<br>Tahun: {row['year']}<br>Indeks Kualitas Udara: {row['kualitas_udara']}<br>Rata-Rata PM10/Tahun: {row['aqi_pm10']}"

    folium.Marker(
        location=[row['latitude'], row['longitude']],
        tooltip=row['station'],
        popup=folium.Popup(popup_text, max_width=200),
        icon=folium.Icon(color=color)
    ).add_to(m)

m
```
